```python
#understanding numpy
import numpy as np
uber_data["MILES_CAT"]=np.where(uber_data['MILES*']>100,"Long Trip","Short Trip")
#where clause
uber_data
```

```python
uber_data['MILES_CAT2'] = np.where(uber_data['MILES*'] < 100, 'Short Trip', np.where(uber_data['MILES*'] <= 200, 'Medium Trip', 'Long Trip'))
uber_data.iloc[450:780]
#where clause with three conditions, if the cond is false moves to false result with another condition(2nd where clause)
```

```python
conditions=[(uber_data['MILES*']<100),(uber_data["MILES*"]>=100)&(uber_data['MILES*']<=200),(uber_data['MILES*']>=200)]
cat=["Short trip","Medium Trip","Long Trip"]
uber_data["MILES_CAT3"]=np.select(conditions,cat,default="Unknown")
#more then one condition statements
```
