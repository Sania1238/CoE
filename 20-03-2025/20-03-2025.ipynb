# Numpy Arange

```python
A = np.arange(100) #Creates an array with 100 elements
print(A)
```

```python
A = np.arange(20,100,3) #Creates an array with 100 elements STARTING with 20 AND takes a JUMP of 3
print(A)
```

# Numpy Slicing

```python
A = np.arange(100)
b=A[3:10]
```

```python
print(b)
```

```python
b[0]=-1200
b #element in index 0 is changed
```

```python
A #The change is visible in main array aswell
```

```python
b=A[3:10].copy() #leaves a copy
b[1]=300
b
```

```python
A #change is not visible in the main array
```

```python
idx = np.argwhere(A==-1200)[0][0]
idx
```

```python
A[idx] = 3 # converting the element in index 3 from -1200 to 3
```

```python
A
```
