```python
import pandas as pd
df=pd.read_csv("Customers.csv")
df
```

```python
#Customer Demographics: Visualize age distribution(hist) and suspending patetrns(scatter plot)
```

```python
import matplotlib.pyplot as plt
a=df['Age']
a
```

```python
plt.hist(a,bins=15,color='skyblue',edgecolor='black')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
b=df['Spending Score (1-100)']
b
```

```python
plt.scatter(a,b,color='orange')
plt.title("Suspension pattern: Age vs Spending score")
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
data=pd.read_csv('amazon.csv.zip')
data
```

```python
data = {
    'product_name': ['Product 1', 'Product 2', 'Product 3'],
    'discounted_price': ['₹399', '₹199', '₹199'], 
    'rating_count': ['24269', '43994', '7928'] 
}
df = pd.DataFrame(data)
df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
df['rating_count'] = pd.to_numeric(df['rating_count'])
df['sales'] = df['rating_count'] * df['discounted_price']

plt.plot(df['product_name'], df['sales'], marker='o', linestyle='-', color='blue')
plt.title('Estimated Monthly Sales')
plt.xlabel('Product')
plt.ylabel('Sales (₹)')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['rating_count'], bins=5, color='green', edgecolor='black')
plt.title('Order Quantity Distribution (based on Rating Count)')
plt.xlabel('Order Quantity (Rating Count)')
plt.ylabel('Frequency')
```

```python
data = {
    'Student Name': ['John', 'Mary', 'Alex', 'Sarah', 'Tom'],
    'Math': [85, 90, 72, 95, 60],
    'Science': [78, 85, 65, 90, 70],
    'English': [92, 88, 78, 85, 65],
    'History': [88, 91, 70, 94, 60]
}
df = pd.DataFrame(data)
df.set_index('Student Name', inplace=True)
subject_averages = df.mean()
subject_averages.plot(kind='bar', color='skyblue', figsize=(10, 6))

plt.title("Average Marks Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10, 6))
df.plot(kind='box', vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen', color='blue'), flierprops=dict(marker='o', markerfacecolor='red', markersize=8))
plt.title("Marks Distribution and Outliers Across Subjects")
plt.xlabel("Marks")
plt.tight_layout()
plt.show()
```
