```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.boxplot(x,y)
```

```python
plt.title("Simple Line plotr")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='purple',linestyle='dotted',linewidth=2,marker='*')
plt.title("Customized Line Plot")
```

```python
#Task-1
x1=[1,2,3,4,5,6]
y1=[3400,5600,7800,4500,9000,7800]
plt.plot(x1,y1,color='green',linestyle='dashed',linewidth=2,marker='o')
plt.title("Savings account over 6-months Line Plot")
plt.xlabel("Months")
plt.ylabel("Savings amount")
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
#create a bar chart
plt.bar(categories,values,color='skyblue')
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
plt.bar(categories,values,color=['skyblue','green','blue','yellow','red'])
```

```python
plt.bar(categories,values,edgecolor='black')
```

```python
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,41,45,56,55,57,68,61,62]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age Distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.show()
```

```python
pl=['Puby','Python','Java','C++']
per = [10,20,30,40]
colors=['gold','lightblue','lightgreen','pink']
plt.pie(per,labels=pl,autopct='%1.1f%%',startangle=90,colors=colors)
plt.title("Programming languages used")
plt.show()
```

```python
x=[5,10,15,20,25]
y=[10,20,25,30,35]
plt.scatter(x,y,color='green',s=50,edgecolor='black')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
