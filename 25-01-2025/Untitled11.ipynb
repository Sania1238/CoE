```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")
df
```

```python
data=pd.read_csv("test.csv")
data
```

```python
import seaborn as sns
sns.barplot(data=df,x='Pclass',y='Survived',errorbar=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
df.columns
```

```python
#is there a relationship between passenger age and fare?
sns.scatterplot(data=df, x="Age", y="Fare",hue='Survived',palette='coolwarm',alpha=0.6)
plt.title("Relationship between Passenger Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
subset=df[['Age','Fare','Pclass','Survived']]
sns.pairplot(subset,hue="Fare",palette="husl")
plt.suptitle('Pairwise Relationships',y=1.02)
plt.show()
```

```python
ncol=df.select_dtypes(include='number')
cmax=ncol.corr()
sns.heatmap(cmax, annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
